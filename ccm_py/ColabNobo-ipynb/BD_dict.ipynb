```
# Төрсөн өдрийн хадгалсан мэдээллийг нь гаргана.
birthdays = {'Зориг': 'Jan 1', 'Чимэг': 'Feb 24', 'Хөлөг': 'Jun 24', 'Хуяг': 'May 15'}

while True:
  print('Нэрийг нь оруул: (хоосон үлдээвэл бүртгэлээс гарна)')
  name = input()
  if name == '':
    break

  if name in birthdays:
    print(name+ '-н төрсөн өдөр бол '  + birthdays[name])
  else:
    print(name + '-н төрсөн өдрийн мэдээлэл байхгүй байна')
    print('Түүний төрсөн өдөр нь хэзээ вэ?')
    bday = input()
    birthdays[name] = bday
    print('Төрсөн өдрийн мэдээлэл шинэчлэгдлээ.')
```
