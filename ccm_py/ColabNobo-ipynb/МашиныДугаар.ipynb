```
carNumbers = []
while True:
  print(str(len(carNumbers) + 1) + '-р машины дугаарыг оруул (Бичихгүй орхивол зогсоно.):')
  number = input()
  if number == '':
    break
  carNumbers = carNumbers + [number] # list concatenation
print('Бүртгэгдсэн машины дугаарууд:')
for number in carNumbers:
  print(' ' + number)
```
