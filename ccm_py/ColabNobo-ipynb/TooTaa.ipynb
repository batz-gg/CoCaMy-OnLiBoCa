```
# Тоо таадаг тоглоом.
import random
secretNumber = random.randint(1, 20)
print('Жүр үрийн Жимсээ би 1-ээс 20 дотор тоо саная. Танд 4 боломж байна.')
 
# Тоглогчоос 5 удаа таамаглалыг нь асууна.
for guessesTaken in range(1, 5):
  print('Таагаарай.')
  guess = int(input())
  
  if guess < secretNumber:
      print('Арай багадууллаа.')
  elif guess > secretNumber:
      print('Арай ихдүүллээ.')
  else:
      break     # Энэ тохиолдолд зөв таасан байна!
if guess == secretNumber:
  print('Баяр хүргэе. Та ' + str(guessesTaken) + ' таамаглалаар таалаа!')
  print('Хэрэглэгч та "500 ЖҮРХЭН ОНОО"-г амжилттай олзлолоо!')
else:
  print('Хараахан биш байна. Би ' + str(secretNumber) + ' гэсэн тоог санасан юм.')
  print('Танд хэрэглэгчийн боловсролд зориулсан дараах мэдээллийг хүргэж байна.')
 
#Бүтээсэн: Г.Батзориг. 2021 он.
```
